```python
import os
from tokenizers import Tokenizer, models, pre_tokenizers, decoders, trainers
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from tqdm.auto import tqdm
```

```python
CLEAN_CORPUS_FILE = 'ces_wikipedia_2021_1M/cleaned_czech_corpus_for_training.txt'
TOKENIZERS_DIR = 'tokenizers_cs'
```

```python
def train_bpe_tokenizer(corpus_file, vocab_size, output_dir):
    tokenizer = Tokenizer(models.BPE())
    tokenizer.pre_tokenizer = pre_tokenizers.Metaspace()
    tokenizer.decoder = decoders.Metaspace()
    trainer = trainers.BpeTrainer(
        vocab_size=vocab_size,
        show_progress=True,
        min_frequency=2,
        special_tokens=[
            "[UNK]",
            "[CLS]",
            "[SEP]",
            "[PAD]",
            "[MASK]",
        ]
    )
    tokenizer.train(files=[corpus_file], trainer=trainer)
    save_path = os.path.join(output_dir, f"bpe-vocab_{vocab_size}.json")
    tokenizer.save(save_path)
    return tokenizer
```

```python
train_bpe_tokenizer(corpus_file=CLEAN_CORPUS_FILE, vocab_size=4000, output_dir=TOKENIZERS_DIR)
```

```python
train_bpe_tokenizer(corpus_file=CLEAN_CORPUS_FILE, vocab_size=8000, output_dir=TOKENIZERS_DIR)
```

```python
VOCAB_SIZES = [4000, 8000, 16000, 32000, 52000]
```

```python
trained_tokenizers = {}
for size in VOCAB_SIZES:
    path = os.path.join(TOKENIZERS_DIR, f"bpe-vocab_{size}.json")
    trained_tokenizers[size] = Tokenizer.from_file(path)
```

```python
for size in VOCAB_SIZES:
    tokenizer = trained_tokenizers[size]
    vocab = tokenizer.get_vocab()
    sorted_vocab = sorted(vocab.items(), key=lambda item: item[1])

    print(f"--- Analysis of {size} Vocabulary ---")
    print(f"Total vocab size: {len(sorted_vocab)}")

    print("\nFirst 15 tokens (foundational characters and special tokens):")
    for token, token_id in sorted_vocab[:15]:
        print(f"ID: {token_id}, Token: '{token}'")

    print("\nLast 15 tokens (more complex/less frequent subwords):")
    for token, token_id in sorted_vocab[-15:]:
        print(f"ID: {token_id}, Token: '{token}'")
```

```python
with open(CLEAN_CORPUS_FILE, 'r', encoding='utf-8') as f:
    corpus_sample = [next(f) for _ in range(20000)]
```

```python
token_lengths = {}

for size, tokenizer in trained_tokenizers.items():
    all_tokens = []
    encoded_batch = tokenizer.encode_batch(corpus_sample)
    for encoding in tqdm(encoded_batch):
        all_tokens.extend(encoding.tokens)
    
    token_lengths[size] = [len(token) for token in all_tokens]
```

```python
plt.figure(figsize=(15, 8))
sns.set_palette("viridis")

for size in VOCAB_SIZES:
    sns.kdeplot(token_lengths[size], label=f'Vocab Size {size}', clip=(0, 20), fill=True, alpha=0.3)

plt.title('Distribution of Token Lengths by Vocabulary Size', fontsize=16)
plt.xlabel('Length of Token (characters)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
